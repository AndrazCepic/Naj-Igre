## Nastavlanje podatkovne strukture in okolja

Vključitev knjižic, ki jih potrebujemo za analizo podatkov. Uporabljali smo knjižici pandas in matplotlib.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

Prebermo osnovne podatke v objekt data, ki ga bomo od zdaj naprej uporabljali za analizo.
Podatke o žanrih pa shranimo v objekt genres.

```python
data = pd.read_csv("data/games.csv")
genres = pd.read_csv("data/genres.csv")
```

Pomnožimo polje 'user_score' z 10, da pridemo na isto skalo kot polje 'metascore'. 

Ne dobimo izjem, torej so vsi vnosi številski kot morajo biti.

```python
data['user_score'] = data['user_score'] * 10.0
```

Odstranimo neveljavne vnose.

Ti so:
* metascore in user_score nista števili
* Imamo NA / NaN vnose

```python
data = data[data['metascore'].astype(str).str.isdigit()].dropna()
```

Polje 'metascore' spremenimo v tip float za računanje.

```python
data['metascore'] = data['metascore'].apply(lambda x: float(x))
```

Dodamo stolpec za leto izida posamezne igre, ki smo ga razbrali iz datuma.
Ogledamo si trenuten izgled tabele podatkov preden začnemo z analizo.

```python
data['year'] = data.date.astype(str).str.slice(start=-4)
data.sort_values('year')
```

## Metascore proti ocenam uporabnikov

Tu primerjamo ocene profesionalnih ocenjevalcev, ki so shranjene v stolpcu 'metascore', z ocenam splošnih uporabnikov, ki so v stolpcu 'user_score'.

```python

data.sort_values('metascore').plot.scatter('metascore', 'user_score', figsize=(20,5))
```

Na grafu se hitro vidi, da so ocene uporabnikov skoraj vedno zelo visoke in so same razlike bolj subtilne, vendar pa vidimo, da so ocene kritikov bolj razpršene. To pa ni začudojoče, saj je kritikov veliko manj kot je splošnih uporabnikov.
Zato bomo zožili interval ocen uporabnikov, da dobimo natančnejšo sliko.

```python
data[data['user_score'] >= 70].sort_values(
    'metascore'
).plot.scatter(
    'metascore', 'user_score', figsize=(20,5)
)
```

Zdaj bolje vidimo, da se ocene uporabnikov in kritikov sploh ne ujemjo dobro, saj hitro vidimo, da imajo na primer dobre ocene kritikov cel spekter slabih ocen do dobrih ocen uporabnikov.

Tedaj smo torej potrdili hipotezo, da se ocene uporabnikov vidno ne ujemajo z ocenami kritikov, saj bi drugače v nekem približku imeli neko funkcijo, ki bi oisovala razmerje teh dveh.

### Posledice:
Ker se ocene ne ujemajo dobro, bomo od zdaj naprej podatke analizirali posebej z metriko ocen uporabnikov in z metriko ocen kritikov, ponekod pa bomo še posebej prikazali rezultate povprečij obeh skupaj.

## Žanri iger

Tabeli data bomo pridružili(join) tabelo genres in analizirali uspešnost različnih žanrov. Analizirali bomo spet posebej za ocene kritikov in ocene uporabnikov.

Prvo bomo določili število žanrov in povprečno število iger na žanr, da bomo lahko ignorirali anomalije. Ta račun pa moramo narediti na zddruženih tabelah, saj smo iz data na začetku odstranili neveljavne igre, pri žanrih pa ne.

```python
data_g = data.merge(genres)
data_g[1000:10000:1000]
```

```python
data_g.groupby('genre').size().sort_values(ascending=False)
```

Vidimo, da imamo 158 žanrov, zdaj pa izračunajmo še povprečno število iger na žanr.

```python
avrg_num_per_genre = int(data_g.groupby('genre').size().sum() /  158)
avrg_num_per_genre
```

Popravimo tabelo žanrov, tako da vzamemo samo žanre, ki imajo vsaj povprečno mnogo iger.

```python
genres_size = genres.groupby('genre').size().reset_index()
genres_size = genres_size[genres_size[0] >= 99]
genres_fixed = genres.merge(genres_size)
data_g = data.merge(genres_fixed)
del data_g[0]
data_g[1000:10000:1000]
```

```python
data_g_m = data_g[['genre', 'metascore']]
data_g_u = data_g[['genre', 'user_score']]
```

### Metascore

```python
data_g_m.groupby(
    'genre'
).mean(
).sort_values(
    'metascore', ascending=False
).head(5)
```

Najkvalitetnejši žanr iz strani kritikov je potemtakem Open-World. To sicer ovrže našo hipotezo, vendar je pa res, da veliko RPG iger spada tudi pod Open-World žanr in obratno.

Pa izpišimo še katere igre so iz tega žanra:

```python
open_world = data_g[data_g['genre'].str.match('^Open-World$')]
```

Po Metascore ocenah:

```python
open_world.sort_values(
    'metascore', ascending=False
).head(5).style.hide_index()
```

Po ocenah uporabnikov:

```python
open_world.sort_values(
    'user_score', ascending=False
).head(5).style.hide_index()
```

### Uporabniki

```python
data_g_u.groupby(
    'genre'
).mean(
).sort_values(
    'user_score', ascending=False
).head(5)
```

Tu pa se končno naša hipoteza ujema z rezultatom! Čeprav je ta rezulat bolj specifično Console-style RPG, bomo predpostavili, da to spada pod RPG in sklepali, da se ujema.

Izpišimo še par iger tega žanra:

```python
rpg = data_g[data_g['genre'].str.match('^Console-style RPG$')]
```

Po Metascore:

```python
rpg.sort_values(
    'metascore', ascending=False
).head(5).style.hide_index()
```

Po uporabnikih:

```python
rpg.sort_values(
    'user_score', ascending=False
).head(5).style.hide_index()
```

Pri tem delu smo potrdili našo hipotezo vsaj iz strani uporabnikov, zdaj pa preverimo, če se morda ujema na povprečju obeh.

Za to bomo morali prvo "normalizirati" ocene uporabnikov, saj od začetka te analize vemo, da se večina giblje v intervalu od 75 do 100. To bomo storili kar linearno.
Ko bomo imeli primerljive vrednosti za uporabnike in metascore, bomo lahko oceni kar sešteli in primerjali seštevke, saj deljenje ne bo nič spremenilo.

```python
data_g.sort_values('user_score', ascending=False).head(1).style.hide_index()
```

Razberemo, da je vrh ocen uporabnikov 94.

```python
high = 94
low = 75

data_g_new = data_g.copy()

data_g_new['user_score'] = data_g['user_score'].apply(lambda x: (high / (high - low)) * (x - low))

# Opustimo vnose, ki so pristali pod 0
data_g_new = data_g_new[data_g_new['user_score'] >= 0.0]

data_g_new.sort_values('user_score', ascending=False)
```

Povprečimo stolpca:

```python
data_g_new['general_score'] = (data_g_new['metascore'] + data_g_new['user_score']) / 2
```

Zdaj pa analizirajmo kateri žanr je povprečno najboljši za te povprečne ocene.

```python
general_gen = data_g_new[['genre', 'general_score']]
general_gen.groupby(
    'genre'
).mean(
).sort_values(
    'general_score', ascending=False
).head(5)
```

In spet se naša hipoteza ujema tudi v povprečju ocen kritik in uporabnikov! Še več, celo specifičen žanr je enak, in sicer Console-style RPG!

Da povzamemo, smo enkrat za spremembo potrdili hipotezo, da je RPG žanr najkvalitetnejši iz strani ocen uporabnikov in povprečij obeh vrst ocen. Za ocene kritikov posebej je pa zmagal žanr Open-World.

## Povezava ESRB ozančbe s kvaliteto iger

Vsaka uradno izdana igra dobi t.i. Entertainment Software Rating Board(ESRB) oznako, ki zaznamuje kakšni publiki je igra namenjena, namreč obstaja na primer veliko nasilnih iger, ki niso najbolj primerne za otroke.
Zanima nas, ali obstaja kakšna korelacija med temi oznakami in kvalito iger. Lahko bi na primer intuitivno sklepali, da so igre, ki so primerne za vse starosti, manj "resne" in morda niso tako visoko ocenje.

To seveda najbolje storimo grafično.

### Metascore

Še graf povprečij:

Iz grafa povprečij je razvidno, da se v povprečju ocene kritiov ne razlikujejo glede na ESRB oznako. S tem potrjuejmo hipotezo kar se tiče ocen kritikov.

## Leto iger!

Za konec pa še poizvedimo katero leto v našem vzorcu je bilo najboljše za igre po ocenah in katero je imelo največ proizvoda.

Da dobimo res eno samo leto iger, dajmo spet delati s povprečjem ocen kritikov in ocen uporabnikov, na kar moramo ponoviti postopek kot prej na tabeli data.

```python
high = 94
low = 75

data_new = data.copy()

data_new['user_score'] = data['user_score'].apply(lambda x: (high / (high - low)) * (x - low))

# Opustimo vnose, ki so pristali pod 0
data_new = data_new[data_new['user_score'] >= 0.0]
data_new['general_score'] = (data_new['metascore'] + data_new['user_score']) / 2
data_new.sort_values('general_score', ascending=False).head(10)
```

### Najboljše leto

```python
naj_leta = data_new[
    ['year', 'general_score']
].groupby(
    'year'
).mean(
).sort_values(
    'general_score', ascending=False
)

naj_leta.head(5)
```

```python
naj_leta.sort_values('year').plot()
```

Zanimivo. Presenetljivo povprečna kvaliteta iger po letih precej upada, naše leto iger je pa 1998!

Igre tistega leta:

```python
best_year = data_new[data_new['year'].astype(int) == 1998]
best_year.sort_values(
    'general_score', ascending=False
).head(10).style.hide_index()
```

Za primerjavo prikažimo še igre vseh let urejene po povprečju ocen.

```python
data_new.sort_values(
    'general_score', ascending=False
).head(10).style.hide_index()
```

Opazimo, da se The Legend of Zelda: Ocarina of Time, ki je zmagovalka leta 1998, pojavi tudi na seznamu vseh iger na sedmem mestu!

### Najbolj produktivno leto

```python
prolific_year = data_new.groupby(
    'year'
).size()
prolific_year.sort_values(ascending=False).head(10)
```

Prikažimo še stopničsti diagram produktivnosti:

```python
prolific_year = prolific_year.reset_index()
prolific_year['number_of_games'] = prolific_year[0]
del prolific_year[0]
prolific_year.sort_values('year').plot.bar('year', 'number_of_games', figsize=(20, 5))
```

Največ iger je bilo torej izdanih leta 2003, na diagramu pa se vidi tudi, kako je od leta 1999 do 2003 zelo hitro naraslo število izdanih iger na leto, kar morda implicira, da je takrat močno narasla popularnost video iger.
